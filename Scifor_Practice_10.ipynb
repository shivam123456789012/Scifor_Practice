```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv("kc_house_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df.drop(['id','date'],axis=1)
```

```python
df.head()
```

```python
X=df.iloc[:,1:]
X[:5]
```

```python
Y=df.iloc[:,0]
Y[:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state=32)
```

```python
reg = RandomForestRegressor(n_estimators=30,random_state=0,oob_score=True) 
reg.fit(X_train, Y_train)
```

```python
Y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print("r2_score:",r2_score(Y_test,Y_pred))
```

```python
print("mse:",mean_squared_error(Y_test,Y_pred))
```

```python
oob_score = reg.oob_score_
print("Out-of-Bag Score:",oob_score)
```

```python
from sklearn.tree import plot_tree
```

```python
tree=reg.estimators_[0]
plt.figure(figsize=(20,10))
plot_tree(tree,feature_names=df.columns.tolist(),filled=True)
plt.title("tree")
plt.show()
```
